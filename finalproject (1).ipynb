Calculate the Total of marks

```python

subjects=['Math','Science','English','Programming','Statistics']
df['total']=df[subjects].sum(axis=1)
df[['Student_ID','Department','total']].head(7)
```

Average marks for each student

```python
df['average']=df[subjects].mean(axis=1)
df[['Student_ID','average']].head(8)
```

Calculating the average marks for each subjects

```python
df[['Math','English','Statistics','Programming','Science']].mean()
```

Calculating Grade

```python
def grade(avg):
 if avg>=90:
  return'A'
 elif avg>=80:
  return'B'
 elif avg>=70:
  return'C'
 elif avg>=60:
  return'D'
 elif avg>=50:
  return'E'
 else:
  return'F'
df['Grade']=df['average'].apply(grade)
df[['Student_ID','average','Grade']].head(10)
```

Highest marks in each subject

```python
df[['Math','English','Programming','Statistics','Science']].max()
```

Lowest mark in each subject

```python
df[['Math','English','Programming','Statistics','Science']].min()
```

Marks greater than 90 in Statistics

```python
df[df['Statistics']>90]
```

Average Marks of Each Subject in Bar graph

```python
import matplotlib.pyplot as plt

subject_average = df[['Math', 'English', 'Science','Programming','Statistics']].mean()

subject_average.plot(kind='bar',color=['lightblue','pink','violet','red','lightgreen'])
plt.title("Average Marks of Each Subject")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.show()
```

Department-wise Student Distribution

```python
import matplotlib.pyplot as plt

df['Department'].value_counts().plot(kind='pie',autopct='%1.1f%%',
colors = ['skyblue', 'lightgreen', 'orange', 'pink', 'yellow', 'violet'])

plt.title('Department-wise Student Distribution')
plt.ylabel('')
plt.show()
```

Percentage for each students

```python
df['Percentage'] = df[['Math', 'Science', 'English', 'Programming', 'Statistics']].mean(axis=1)
df.head()
```

Age-wise Academic Performance

```python
import matplotlib.pyplot as plt

age_performance = df.groupby('Age')['Percentage'].mean()

age_performance.plot(kind='bar')
plt.title('Age-wise Academic Performance')
plt.xlabel('Age')
plt.ylabel('Average Percentage')
plt.show()
```
